regex: '^\s*$'
- ^: 문자열 시작
- \s*: 공백이 없거나 있거나
- $: 문장의 끝

```python
df['산업분류2'] = df['산업분류2'].replace(r'^\s*$', np.NaN, regex=True) #regex 이용하여 빈칸인 속성 지정하고 NaN 결측치로 대체.
df.iloc[30:40,0:3]
```

- 산업분류2 : replace(), regex 적용.

```python
df['산업분류2']=df['산업분류2'].str.replace(r'1\)','',regex = True)   #regex 정규표현식으로 '1)' 부분 삭제
df.iloc[30:40,0:3]
```

for문, if문을 이용하여 산업분류2,3 모두 NaN인 행의 인덱스 추출

- 하나의 값을 가져와서 결측치 여부를 확인하기 위해 pd.isna(df['col_name'][i]) 형식으로 작성.

##### melt() 이용하여 wide format을 long format으로 변환 & 새 컬럼 'coefficient', 'value' 생성

```python
df.drop([0],axis=0,inplace=True) #첫 행(연도) 일단 삭제. 추후 'year' 열로 추가.
df.head()
```

기준열: ['산업분류1','산업분류2','산업분류3']  \
새로운 열: coefficient, value
  - coefficient: 기존 df에서 column name을 변환
  - value: 기존 df에서 속성값을 변환

```python
df = df.melt(
    id_vars=['산업분류1','산업분류2','산업분류3'],   #그룹식별자 column
    var_name='coefficient',                        #기존의 column name을 'coefficient' 열로 변환
    value_name='value'                             
)
df
```

coefficient 컬럼의 "Unnamed: 숫자"를 앞의 값으로 대체

```python
df.iloc[170:180,:]
```

```python
df = df.assign(
    coefficient =lambda x:
    x.coefficient                                      #기존의 coefficient 컬럼에서
    .str.replace("Unnamed: \\d+","",regex = True)      #regex를 이용하여 'Unnamed: 숫자'를 ''로 대체
    .replace("",np.nan)                                #''를 NaN으로 대체
    .ffill()                                           #NaN을 앞의 속성값으로 대체
)
df.iloc[170:180,:]                 
```

```python
#coefficient열이 제대로 대체되었는지 확인
print(df['coefficient'].value_counts())
df['coefficient'].info
```

year 열 생성하고 2015-2018 반복 입력
- long format의 형태를 보면 wide format의 8개 열이 세로로 붙은 것을 알 수 있다.
- year 열은 2015,2016,2017,2018,2015,2016,2017,2018이 연속적으로 44개씩 반복되어야 한다.

```python
df = df.assign(
    year = lambda x:
        np.repeat([2015,2016,2017,2018,2015,2016,2017,2018],44)
)
df
```

```python
df['year'].value_counts()    #year열 data 빈도 확인
```
